#Ex2.**Simple Optimization**

```python
import numpy as np
def find_minimum(f, min_value, num_iteration, step):
    delta_x = 1e-6
    for i in range(num_iteration):
        dx = (f(min_value + delta_x / 2) - f(min_value - delta_x/2)) / delta_x
        dx_sign = np.sign(dx)
        if dx_sign == 0:
            break
        else:
            min_value = min_value - dx_sign*step
        print(f'[INFO] Iterattion ({i+1}/{num_iteration}): New x = {min_value}')
    return min_value
```

```python
import random
def f(x):
    return 3*x**4 - 4*x**2 - 6*x - 3

x = random.uniform(-10, 10)
print('Initial x: ', x)

min_value = find_minimum(f=f, min_value=x, num_iteration=2000, step=0.01)
print(min_value)
```

#Ex3.**Simple Optimization (Chain Rule)**

##3.1.**Find minimum function**

```python
def find_minimum(f, g, min_value, num_iteration, step):
    delta_x = 1e-6
    delta_f = 1e-6
    for _ in range(num_iteration):
        df_dx = (f(min_value + delta_x/2) - f(min_value - delta_x/2)) / delta_x
        fx = f(min_value)
        dg_df = (g(fx + delta_f / 2) - g(fx - delta_f / 2)) / delta_f
        dg_dx = dg_df * df_dx
        dg_dx_sign = np.sign(dg_dx)

        if dg_dx_sign == 0:
            break
        else:
            min_value = min_value - dg_dx_sign*step
    return min_value
```

```python
import random
def f(x):
    return 2* x **3 - 3* x **2 + 4* x

def g(x):
    return x **2 - 10* x + 2

x = random . uniform ( -2 , 2)
print (" initial x: ", x )

min_value = find_minimum (f =f , g=g , min_value=x , num_iteration=40 , step=0.1)
print (min_value)
```
